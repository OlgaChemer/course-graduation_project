```python
# минимальный, максимальный, средний, медианный возраст студентов

# сначала удаляем пустые значения в столбце
df_students.dropna(inplace = True)

# далее пишем функцию для рассчета возраста на сегоднешний день

def age(x):                                                      

    return (dt.now() - dt.strptime(str(x).replace('-','/'),"%Y/%m/%d")).days//365

# создаем новый столбец с возрастом
df_students['age'] = df_students.loc[:,'birthday'].map(lambda x: age(x))  

# считаем по новому столбцу минимальный, максимальный, средний, медианный возраст студентов 
print('Минимальный возраст', df_students['age'].min(), '\nМаксимальный возраст', 
      df_students['age'].max(), '\nСредний возраст', df_students['age'].mean(), 
      '\nМедианный возраст', df_students['age'].median())
```

```python
# минимальный, максимальный, средний, медианный возраст студентов на каждом курсе

# Сначала переименуем столбец с идентификатором студента для дальнешего объединения столбцов
df_students = df_students.rename(columns = {'id':'student_id'})

# объединяем датасеты по id студента
df_age = df_students.merge(df_progresses, on='student_id')
df_age = df_age.merge(df_courses, on='course_id')

# группируем датасет по всем курсам и создаем новые стобцы, агрегируя возраст
df_age.groupby('title').agg(min_age = ('age', 'min'),
                                max_age = ('age', 'max'),
                                mean_age = ('age', 'mean'),
                                median_age = ('age', 'median'))
```

Задание 2.1 
Рассчитать прирост студентов на каждом курсе в каждом месяце за всю историю (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно). Считать дату начала прохождения курса студентом по дате начала первой домашней работы.

```python
# Меняем формат столбца с датай с object на datetime64
df_joint['start_date'] = pd.to_datetime(df_joint['start_date'])

# создаем столбец с датай в формате год-месяц, удаляя пустые значения
df_joint['month_year'] = df_joint['start_date'].dropna().apply(lambda x: x.strftime('%Y-%m'))

# создаем столбец с датай в формате год-месяц-день
df_joint['start_date_day'] = df_joint['start_date'].apply(lambda x: x.date())
```

```python
# формируем датасет, группируя по столбцам название курса и месяц-год и агрегируем по минимальному модулю

df = df_joint[(df_joint['module_number'].min()) & (df_joint['is_homework'] == True)].groupby('title').agg({'module_number': 'min'}).reset_index()
df
```

```python
# объединяем получившийся датафрейм и и исходный датафрейм по названию курса и номеру модул

df_count_of_work = df_joint.merge(df, left_on=['title', 'module_number'], right_on=['title', 'module_number'])
df_count_of_work
```

## 5. Метрика успеваемости 

&nbsp;&nbsp;&nbsp;&nbsp;*(необязательное задание)*

Иногда студенты берут курсы, которые оказываются для них неподъемными. Это может быть как по причинам недостаточной изначальной подготовки, так и по причинам, связанным с низкой мотивацией студента. Для улучшения качества контента полезно выявить причину. На основании имеющихся данных придумайте метрику успеваемости студента. **Обоснуйте ее.** Выявите таких студентов на каждом курсе, чтобы предоставить продюсерам список проблемных студентов. 

```python
# Для анализа успеваемости студентов будем анализировать датасет по выполненым домашним заданиям. 
# Не выполненные задания в столбце status имеют значения fail. Тогда будем считать, что студенты 
# с невыполненными заданиями от 3 и более являются проблемными студентами.

# отфильтруем датасет по тем строкам, где есть домашнее задание
df = df_joint[df_joint.loc[:,'is_homework'] == True]

# выберем строки со значениями столбца status 'fail' и 'start', сгруппируем по курсу, id студента(др и город для 
# доп анализа), и посчитаем количесвто не выполненных заданий
df = df[(df['status'] == 'fail') | (df['status'] == 'start')].groupby([ 
    'title', 'student_id', 'birthday', 'city']).agg(fail_count = ('status','count')).reset_index()

# отсортируем количесвто не выполненных заданий больше 2
df = df[df['fail_count'] > 2]
df
```

```python
# Агрегируем датасет по курсам с подсчетом студентов

df_problem = df.groupby('title').agg(students_count = ('student_id','count')).sort_values('students_count', ascending=False)
df_problem
```
